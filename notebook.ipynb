## 2. Trend Analysis:

Identify trends among the top YouTube streamers. Which categories are the most popular? Is there a correlation between the number of subscribers and the number of likes or comments?

```python
# Selecting only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## Observations

Interestingly, there's no evident correlation between the number of subscribers and likes, nor between subscribers and comments. The "Rank" column shows no meaningful correlation with any other column in the dataset, offering no valuable insights and is hence recommended for removal. However, a notable strong correlation exists between the number of visits and likes. There seems to be a weaker linear relationship between visits and comments, as well as between likes and comments. Additionally, there's no apparent correlation between the number of comments and the count of subscribers.

```python
# Dropping `Rank` from the dataframe 
df.drop(axis=1, columns='Rank', inplace= True) 
```

## 7. Benchmarking:

- Identify streamers with above-average performance in terms of subscribers, visits, likes, and comments.
- Who are the top-performing content creators?

```python
# Calculate average performance metrics
average_metrics = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()
```

```python
# Identify streamers with above-average performance
above_average_streamers = df[
    (df['Suscribers'] > average_metrics['Suscribers']) &
    (df['Visits'] > average_metrics['Visits']) &
    (df['Likes'] > average_metrics['Likes']) &
    (df['Comments'] > average_metrics['Comments'])
]
```

```python
# Find top-performing content creators (e.g., top 10 based on subscribers)
top_performers = df.nlargest(10, 'Suscribers')
```

```python
# Display above-average streamers and top-performing content creators
print("Streamers with Above-Average Performance:")
print(above_average_streamers[['Username', 'Suscribers', 'Visits', 'Likes', 'Comments']])

print("\nTop-Performing Content Creators:")
print(top_performers[['Username', 'Suscribers', 'Visits', 'Likes', 'Comments']])
```

## 8. Content Recommendations:

- Propose a system for enhancing content recommendations to YouTube users based on streamers'
- categories and performance metrics.

```python

# Step 1: User preferences (simulated)
user_preferences = {'Categories': ['Music and Dance', 'Video Games'], 'Suscribers': 1000000, 'Visits': 50000}

# Step 2: Filtering based on user preferences
filtered_streamers = df[
    (df['Categories'].isin(user_preferences['Categories'])) &
    (df['Suscribers'] >= user_preferences['Suscribers']) &
    (df['Visits'] >= user_preferences['Visits'])
]

# Step 3: Sort and recommend top streamers
recommended_streamers = filtered_streamers.sort_values(by=['Suscribers', 'Visits', 'Likes', 'Comments'], ascending=False).head(10)

# Display recommended streamers
print("Recommended Streamers:")
print(recommended_streamers[['Username', 'Categories', 'Suscribers', 'Visits', 'Likes', 'Comments']])
```
